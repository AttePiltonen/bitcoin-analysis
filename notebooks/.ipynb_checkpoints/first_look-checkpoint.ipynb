```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
btc = pd.read_csv('../data/BTC-USD.csv')
btc.head()
```

```python
btc.describe()
```

```python
btc.info()
```

```python
# To look at the daily trend, we only need the index and Close column
btc_daily = btc.drop(columns=['Date', 'Open', 'High', 'Low', 'Adj Close', 'Volume'])
btc_daily
```

```python
btc_daily['Days'] = btc.index + 1
```

```python
btc_daily.plot(x='Days', y='Close',loglog=True)
btc_daily
```

```python
btc_daily.plot(x='Days', y='Close',logy=True)
btc_daily
```

```python
btc_daily.to_csv('../data/btc_daily.csv')
```
